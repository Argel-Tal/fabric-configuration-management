```
def delete_workspace_access(workspace_id_in):
    '''
    ------------------- READ THIS COMMENT --------------------------
    
    Because there is currently no Service Principal support for the Admin APIS `labs.admin.delete_workspace()`, so we are forced to do this component outside of SP contexts.
    Have raised a feature enhancement request on SLL github: https://github.com/microsoft/semantic-link-labs/issues/1104

    If the SP context for this Admin API is added, we can then indent this entire function under 1x `with labs.service_principal_authentication(...)` block
    
    For now it's Fabric tenant perms via PIM only :(


    -------------- THANK YOU FOR YOUR ATTENTION --------------------

    '''

    # https://semantic-link-labs.readthedocs.io/en/stable/sempy_labs.admin.html#sempy_labs.admin.add_user_to_workspace
    labs.admin.add_user_to_workspace(mssparkutils.env.getUserName(),'Admin','User', workspace_id_in)
    # must be a workspace admin to delete a workspace - no Admin APIs for this
    labs.workspace.delete_workspace(workspace_id_in)
    print(f"Deleted workspace: {workspace_id_in}, please ensure it is also deleted from configuration files 'Domain - Workspace mapping' and 'Workspace Access', else these will throw 404 errors")
```
